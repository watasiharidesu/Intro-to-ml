```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import random
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.linear_model import LinearRegression
```

```python
random.seed(123)
```

```python
trfv = trf.values
```

```python
features = trfv[:,0:-1]
```

```python
target = trf.loc[:,"target"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
rf = RandomForestClassifier(n_estimators = 10)
rf.fit(x_train, y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train, y_train)
```

```python
dt_score = accuracy_score(y_test, dt.predict(x_test))
rf_score = accuracy_score(y_test, rf.predict(x_test))
knn_score = accuracy_score(y_test, knn.predict(x_test))
```

```python
print("decision tree score: ", dt_score)
print("randomforest score: ", rf_score)
print("knn score: ", knn_score)
```

```python
x = tr.iloc[:,0:1403]
y = tr.loc[:,"target"]
```

```python
pca = PCA(n_components=9)
pca.fit(x)
pcax = pca.fit_transform(x)
print(pca.explained_variance_ratio_)
```

```python
x_train, x_test, y_train, y_test = train_test_split(pcax, y, test_size=0.2)
```

```python
dt = DecisionTreeClassifier(random_state= 123)
dt.fit(x_train, y_train)
rf = RandomForestClassifier(n_estimators = 10)
rf.fit(x_train, y_train)
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train, y_train)
```

```python
dt_score2 = accuracy_score(y_test, dt.predict(x_test))
rf_score2 = accuracy_score(y_test, rf.predict(x_test))
knn_score2 = accuracy_score(y_test, knn.predict(x_test))
```

```python
print("decision tree score: ", dt_score2)
print("randomforest score: ", rf_score2)
print("knn score: ", knn_score2)
```
